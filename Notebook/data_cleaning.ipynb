```python
df["Order Date"] = pd.to_datetime(df["Order Date"])
df["Ship Date"] = pd.to_datetime(df["Ship Date"])

print("Date columns converted successfully!")
```

```python
print("Order Date Range")

print(df["Order Date"].min())

print(df["Order Date"].max())
```

```python
df["Order Year"] = df["Order Date"].dt.year

df["Order Month"] = df["Order Date"].dt.month_name()

df["Order Quarter"] = df["Order Date"].dt.quarter

df["Processing Days"] = (
    df["Ship Date"] - df["Order Date"]
).dt.days

df["Profit Margin (%)"] = (
    df["Profit"] / df["Sales"]
) * 100

df.head()
```

```python
print("Total Sales")

print(round(df["Sales"].sum(),2))
```

```python
print("Total Profit")

print(round(df["Profit"].sum(),2))
```

```python
print("Total Orders")

print(df["Order ID"].nunique())
```

```python
print("Sales by Region")
sales_region = df.groupby("Region")["Sales"].sum().sort_values(ascending=False)

sales_region
```

```python
print("Profit by Region")
profit_region = df.groupby("Region")["Profit"].sum().sort_values(ascending=False)

profit_region
```

```python
print("Sales by Category")
category_sales = df.groupby("Category")["Sales"].sum().sort_values(ascending=False)

category_sales
```

```python
print("Profit by Category")
category_profit = df.groupby("Category")["Profit"].sum().sort_values(ascending=False)

category_profit
```

```python
print("Sales by Segment")
segment_sales = df.groupby("Segment")["Sales"].sum().sort_values(ascending=False)

segment_sales
```

```python
print("Top 10 Customers")
top_customers = df.groupby("Customer Name")["Sales"].sum().sort_values(ascending=False).head(10)

top_customers
```

```python
print("Top 10 Products")
top_products = df.groupby("Product Name")["Sales"].sum().sort_values(ascending=False).head(10)

top_products
```

```python
print("Monthly Sales Trend")
monthly_sales = df.groupby("Order Month")["Sales"].sum()

monthly_sales
```

```python
print("State-wise Sales")
state_sales = df.groupby("State/Province")["Sales"].sum().sort_values(ascending=False)

state_sales.head(10)
```

```python
print("Ship Mode Analysis")
ship_mode = df.groupby("Ship Mode")["Sales"].sum()

ship_mode
```

```python
print("Discount Analysis")
discount = df.groupby("Discount")["Profit"].mean()

discount
```

```python
print("Correlation")
df[["Sales","Profit","Quantity","Discount"]].corr()
```
